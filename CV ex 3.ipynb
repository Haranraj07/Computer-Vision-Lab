```python
import cv2
import matplotlib.pyplot as plt

# Load the pre-trained Haar Cascade Classifier for face detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Load the image in which faces need to be detected
image = cv2.imread('batman.webp')  # replace with the path to your image

# Convert the image to grayscale (required for face detection)
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Detect faces in the image
faces = face_cascade.detectMultiScale(gray_image, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

# Draw rectangles around the detected faces
image_with_faces = image.copy()  # make a copy of the image to draw faces on
for (x, y, w, h) in faces:
    cv2.rectangle(image_with_faces, (x, y), (x+w, y+h), (255, 0, 0), 2)

# Convert the images from BGR (OpenCV default) to RGB for display with matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_with_faces_rgb = cv2.cvtColor(image_with_faces, cv2.COLOR_BGR2RGB)

# Create subplots to show the original image and face-detected image
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Display the original image
axes[0].imshow(image_rgb)
axes[0].set_title("Original Image")
axes[0].axis('off')  # Hide axis for better presentation

# Display the image with face detection
axes[1].imshow(image_with_faces_rgb)
axes[1].set_title("Face Detected")
axes[1].axis('off')  # Hide axis for better presentation

# Show the plot
plt.tight_layout()
plt.show()
```
